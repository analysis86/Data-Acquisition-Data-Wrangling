```python
cols=['season','hr','holiday','weekday','weathersit','temp','atemp','hum','windspeed','casual','registered','cnt']
final_dataset[cols]=final_dataset[cols].apply(pd.to_numeric)
```

```python
final_dataset
```

```python
def find_outliers(df,cols):
    a1=df[cols].quantile(.30)
    b1=df[cols].quantile(.90)
    ab=b1-a1
    min_n=a1-1.5*ab
    max_n=b1+1.5*ab
    outliers_index=df.index[(df[cols]< min_n) | (df[cols]< max_n)]
    return outliers_index
```

```python
find_outliers(final_dataset,'instant')
```

```python
find_outliers(final_dataset,'dteday')
```

```python
find_outliers(final_dataset,'hr')
```

```python
find_outliers(final_dataset,'atemp')
```
